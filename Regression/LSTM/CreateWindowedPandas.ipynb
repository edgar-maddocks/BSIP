```python
import datetime
import pandas as pd
df = pd.read_csv("MSFT.csv")
df = df[["Date", "Close"]]
def windowed_dataframe(df, window_size):
    for i in range(window_size + 1):
        shift = window_size - i
        df[f"Target - {shift}"] = df["Close"].shift(shift)
    del df["Close"]
    df = df.iloc[window_size:]
    df = df.reset_index()
    del df["index"]
                
    return df

df = windowed_dataframe(df, 3)
df
```
